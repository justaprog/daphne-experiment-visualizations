```python
# create a matrix of size nxm filled with 1s and save it to a matrix_m.mtx file in Matrix Market format.
import numpy as np
from scipy.io import mmwrite

n = 128
m = 128
matrix = np.ones((n, m))
# change last column to 0s
matrix[:, -1] = 0
mmwrite('../data/matrix_m.mtx', matrix)

# write metadata file as matrix_m.mtx.data with the following content:
# {"numCols":,"numRows":,"valueType":"f64", "numNonZeros": }
metadata = {
    "numCols": m,
    "numRows": n,
    "valueType": "f64",
    "numNonZeros": int(np.sum(matrix != 0))
}
import json
with open('../data/matrix_m.mtx.data', 'w') as f:
    json.dump(metadata, f, indent=2)

```
