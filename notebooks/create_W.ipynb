```python
# create a matrix of size nxm filled with 1s (except last col with 0s)and save it to a matrix_m.mtx file in Matrix Market format.
import numpy as np
from scipy.io import mmwrite

n = 66
m = 66

all_pairs = [(66,66), (822,822), (35,35), (800,800), (2000,2000), (115,115), (185,185), (343,343), (512,512), (363,363)]
for n, m in all_pairs:
    matrix = np.ones((n, m))
    # change last column to 0s
    matrix[:, -1] = 0
    output_file = f'../data/w_{n}_{m}.mtx'
    mmwrite(output_file, matrix)

    # write metadata file as matrix_m.mtx.data with the following content:
    # {"numCols":,"numRows":,"valueType":"f64", "numNonZeros": }
    metadata = {
        "numCols": m,
        "numRows": n,
        "valueType": "f64",
        "numNonZeros": int(np.sum(matrix != 0))
    }
    import json
    with open(f'../data/w_{n}_{m}.mtx.meta', 'w') as f:
        json.dump(metadata, f, indent=2)
```
